```python
# Pictures

## We need to install and import the package that handles the images:
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Now, we load the image from the file:

myImage = Image.open("./flower_square.jpg")
plt.figure(figsize=(9, 6))
plt.imshow(myImage)
```

```python
## We will convert the image into a NumPy array.

imgmat = np.array(myImage)
plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```

```python
print(type(myImage))
print(type(imgmat))
print(imgmat.shape)
```

```python
## We can see that these variables have different types but when we plot them, they look exactly the same.

## In NumPy, we have a 3-D array. The first 2 dimensions are 900x900. 
## This represents the size of the picture. The last dimension is 3, therefore we have one layer for red, one for blue and one for green. 
## Each value represents the intensity of a particular color in that specific position.


## Representation of images in Numpy (using NumPy arrays) allows us to do simple changes to these images using mathematical operations.

## Let's cover the corners of the picture with black to make it round:
```

```python
imgmat = np.array(myImage)
# storing the shape of picture into the variables
lx, ly, lz = imgmat.shape

X,Y = np.ogrid[0:lx, 0:ly]
print(X.shape)
print(Y.shape)

mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4

# Where mask is True, insert black color
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0

plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```
